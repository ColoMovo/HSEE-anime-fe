# 高考加油视频 - Remotion 云端渲染 (Google Colab)

使用此 Notebook，你可以在 Google 的免费云服务器上渲染 Remotion 视频，避免占用你的本地电脑资源。

**步骤：**
1. 将整个项目文件夹（包含 `bgm-analysis.json`, `Last Hope - Victor Cooper.flac`, `remotion` 文件夹等）压缩为 `project.zip`
2. 在左侧边栏点击📁文件图标，将 `project.zip` 上传到 Colab 环境。
3. 依次点击下方每个代码块左侧的 ▶️ 播放按钮运行代码。

```python
# 1. 解压项目文件并配置 Node.js 20 环境
!unzip -q project.zip -d project
!curl -fsSL https://deb.nodesource.com/setup_20.x | bash -
!apt-get install -y nodejs ffmpeg libnss3 libatk1.0-0 libatk-bridge2.0-0 libcups2 libdrm2 libxkbcommon0 libxcomposite1 libxdamage1 libxfixes3 libxrandr2 libgbm1 libasound2
```

```python
# 2. 安装项目依赖
%cd /content/project
!npm install
%cd remotion
!npm install
```

```python
# 3. 启动本地预览服务器并开始渲染视频
import subprocess
import time

print("启动预览服务器...")
server_process = subprocess.Popen(["npm", "start"], cwd="/content/project")
time.sleep(5) # 等待服务器启动

print("开始渲染视频...")
!npx remotion render Root.tsx gaokao-video --output=../out.mp4 --concurrency=4

print("渲染完成！正在关闭服务器...")
server_process.terminate()
```

### 下载视频
渲染完成后，在左侧文件浏览器中打开 `project` 文件夹，找到 `out.mp4`，右键点击并选择 **Download (下载)** 即可保存到你的电脑上！
